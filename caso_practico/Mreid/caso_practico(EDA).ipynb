```python
exchange_rates_df = pd.read_csv('exchange_rates.csv')
exchange_rates_df.head()
```

```python
merged_df = pd.merge(df, exchange_rates_df, on='year')
merged_df.tail()
```

```python
merged_df
```

```python
#example of the idea 
if 'FixedassetsthUSD' in merged_df.columns:
    merged_df['FixedassetsEUR'] = merged_df['FixedassetsthUSD'] * merged_df['exchange_rate']
print(merged_df[['year', 'FixedassetsthUSD', 'FixedassetsEUR']])
```

```python
columns_to_convert = [
    'FixedassetsthUSD', 'FixedassetsthUSD_green', 'FixedassetsthUSD_mergers',
    'TotalassetsthUSD', 'TotalassetsthUSD_green', 'TotalassetsthUSD_mergers'
]

for column in columns_to_convert:
    if column in merged_df.columns:
        # Create new column names for the EUR values
        new_column_name = column.replace('USD', 'EUR')
        merged_df[new_column_name] = merged_df[column] * merged_df['exchange_rate']

# Display the first few rows to check the new columns (optional)
merged_df
```
